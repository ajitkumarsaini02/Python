```python
class employee:
    def __init__(self, name, id):
        self.id = id
        self.name = name
    def display(self):
        print("ID: %d \nName: %s" % (self.id, self.name))
emp1 = employee("John", 101)
emp2 = employee("David", 102)
emp1.display()
emp2.display()
```

```python
class stack:
    def push(self, st, data):
        st.append(data)

    def pop(self, st):
        if len(st) > 0:
            ans = st.pop()
            print("Removed element from the stack:", ans)
        else:
            print("The stack is Empty")

    def top(self, st):
        if len(st) > 0:
            return st[len(st) - 1]
        return None

    def isEmpty(self, st):
        if len(st) == 0:
            return 1
        else:
            return 0

    def size(self, st):
        return len(st)

st = stack()
s = ["Lokesh", "Diwakar", "Aniket", "Ritik"]

print(s)
print(st.size(s))
st.pop(s)
print("Top element in the stack:", st.top(s))
print(st.isEmpty(s))
```

```python
class Stack:
    def __init__(self):
        self.items = []

    def is_empty(self):
        return len(self.items) == 0

    def push(self, item):
        self.items.append(item)

    def pop(self):
        if not self.is_empty():
            return self.items.pop()
        else:
            return "Stack is empty"
s = Stack()
print("Pushing elements 10, 20, 30")
s.push(10)
s.push(20)
s.push(30)
print("Current Stack:", s.items)
print("\nPopping one element:")
print("Popped:", s.pop())
print("Current Stack:", s.items)
print("\nPopping remaining elements:")
print("Popped:", s.pop())
print("Popped:", s.pop())
print("\nTrying to pop from empty stack:")
print("Popped:", s.pop())
```

```python
class queue:
    def enqueue(self, q, data):
        q.append(data)

    def dequeue(self, q):
        if len(q) > 0:
            ans = q.pop(0)
            print("Removed element from the queue:", ans)
        else:
            print("The queue is Empty")

    def front(self, q):
        if len(q) > 0:
            return q[0]
        return None

    def isEmpty(self, q):
        if len(q) == 0:
            return 1
        else:
            return 0

    def size(self, q):
        return len(q)


q1 = queue()
q = ["Lokesh", "Diwakar", "Aniket", "Ritik"]

print(q)
print(q1.size(q))
q1.dequeue(q)
print("Front element in the queue:", q1.front(q))
print(q1.isEmpty(q))
```

```python
class Queue:
    def __init__(self):
        self.items = []
    def enqueue(self, item):
        self.items.append(item)
    def dequeue(self):
        if len(self.items) == 0:
            return "Queue is empty"
        return self.items.pop(0)
q = Queue()
print("Enqueuing 10, 20, 30")
q.enqueue(10)
q.enqueue(20)
q.enqueue(30)
print("Current Queue:", q.items)
print("\nDequeuing one element:")
print("Dequeued:", q.dequeue())
print("Current Queue:", q.items)
print("\nDequeuing remaining elements:")
print("Dequeued:", q.dequeue())
print("Dequeued:", q.dequeue())
print("\nTrying to dequeue from empty queue:")
print("Dequeued:", q.dequeue())
```
