Q1. A cashier has currency notes of denominations 10, 50 and 100. If the amount to be withdrawn is input through the keyboard in hundreds, find the total number of currency notes of each denomination the cashier will have to give to the withdrawer.

```python
x=int(input("Enter rupees:"))
n=int(x/100)
x=x%100
m=int(x/50)
x=x%50
p=int(x/10)
print(n," 100 note")
print(m," 50 note")
print(p," 10 note")
```
