### üîÑ **Type Conversion**

* **Implicit Conversion:**
  Python automatically converts one data type to another.
  Example:

  ```python
  a = 5
  b = 2.0
  result = a + b  # result becomes 7.0 (int is converted to float)
  ```

* **Explicit Conversion (Type Casting):**
  You manually convert one data type to another using functions like `int()`, `float()`, `str()`, etc.
  Example:

  ```python
  x = "10"
  y = int(x)  # y is now an integer
  ```

---

### üí¨ **Comments in Python**

* **Single-line comment:**
  Use `#` to write single-line comments.
  Example:

  ```python
  # This is a single-line comment
  ```

* **Multi-line comment:**
  Use triple quotes (`'''` or `"""`) for multi-line comments (although technically these are multi-line strings).
  Example:

  ```python
  '''
  This is a multi-line comment
  spanning multiple lines
  '''
  ```

# üìò Notes: String Operation using `split()`

### üîπ What is `split()`?

* `split()` is a **string method** in Python.
* It breaks a string into a **list of substrings** based on a **separator (delimiter)**.
* By default, it splits using **whitespace (spaces, tabs, newlines)**.

---

### üîπ Syntax

```python
string.split(separator, maxsplit)
```

* **separator (optional):** The delimiter where the string should split.

  * Default: whitespace (`' '`, `'\t'`, `'\n'` etc.)
* **maxsplit (optional):** The maximum number of splits to perform.

  * Default: `-1` (no limit, splits at all occurrences).

---

### üîπ Examples

#### 1. Default split (whitespace)

```python
text = "Python is powerful"
result = text.split()
print(result)  
# Output: ['Python', 'is', 'powerful']
```

#### 2. Split using a specific separator

```python
data = "apple,banana,orange"
result = data.split(",")
print(result)  
# Output: ['apple', 'banana', 'orange']
```

#### 3. Using `maxsplit`

```python
text = "one two three four"
result = text.split(" ", 2)
print(result)  
# Output: ['one', 'two', 'three four']
```

#### 4. Splitting with newline `\n`

```python
lines = "Hello\nWorld\nPython"
result = lines.split("\n")
print(result)  
# Output: ['Hello', 'World', 'Python']
```

---

### üîπ Key Points

* Returns a **list of strings**.
* Useful in **parsing text data**, **reading CSV-like strings**, **tokenizing sentences**, etc.
* If the separator is not found, the whole string is returned as a single element in a list.


````python
#write a python function which performs the run lenth encoding for a given string and return the run length encoded string

def run_length_encode(input_string):
    encoded_string = ""
    if not input_string:
        return encoded_string

    count = 1
    for i in range(len(input_string)):
        if i + 1 < len(input_string) and input_string[i] == input_string[i+1]:
            count += 1
        else:
            encoded_string += input_string[i] + str(count)
            count = 1
    return encoded_string
````

## üìå **Tuple in Python**

### üîπ Definition

* A **tuple** is a **collection data type** in Python.
* It is similar to a list but **immutable** (cannot be changed after creation).
* Tuples are ordered, indexed, and can store **heterogeneous** (different) data types.

---

### üîπ Syntax

```python
# Creating a tuple
t1 = (1, 2, 3)
t2 = ("apple", "banana", "cherry")
t3 = (1, "hello", 3.5, True)
t4 = ()              # Empty tuple
t5 = (5,)            # Single element tuple (comma is must)
```

---

### üîπ Characteristics

1. **Ordered** ‚Üí Elements have a fixed position.
2. **Immutable** ‚Üí Elements cannot be modified, added, or removed.
3. **Allow duplicates** ‚Üí Same element can appear multiple times.
4. **Can contain different data types** ‚Üí int, float, string, list, another tuple, etc.

---

### üîπ Accessing Elements

```python
t = (10, 20, 30, 40)

print(t[0])   # 10 (first element)
print(t[-1])  # 40 (last element)
print(t[1:3]) # (20, 30) (slicing)
```

---

### üîπ Tuple Operations

```python
t1 = (1, 2, 3)
t2 = (4, 5)

print(t1 + t2)   # Concatenation ‚Üí (1, 2, 3, 4, 5)
print(t1 * 2)    # Repetition ‚Üí (1, 2, 3, 1, 2, 3)
print(len(t1))   # Length ‚Üí 3
print(2 in t1)   # Membership ‚Üí True
```

---

### üîπ Built-in Functions

```python
t = (5, 2, 8, 5, 10)

print(max(t))     # 10
print(min(t))     # 2
print(sum(t))     # 30
print(t.count(5)) # 2
print(t.index(8)) # 2
```

---

### üîπ Nested Tuple

```python
t = (1, (2, 3), (4, 5, (6, 7)))
print(t[1][0])    # 2
print(t[2][2][1]) # 7
```

---

### üîπ Tuple Packing & Unpacking

```python
# Packing
t = 10, 20, 30

# Unpacking
a, b, c = t
print(a, b, c)  # 10 20 30
```
