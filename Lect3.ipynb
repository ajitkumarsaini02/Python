üîπ 1. Operator Precedence

* Definition: Precedence determines which operator is evaluated first when an expression has more than one operator.

* Higher precedence ‚Üí evaluated earlier.

   int y = 100 / 10 * 2;  


üîπ 2. Operator Associativity

* Definition: Associativity decides the order of evaluation when operators of the same precedence appear together.

* Commonly, operators are left-to-right associative, but some (like assignment =, unary operators, conditional ?:) are right-to-left.

   int a = b = c = 5;

```python

furniture = ["Sofa Set", "Dining Table", "T.V. Stand", "Cupboard"]
cost = [20000, 8500, 4599, 13920]
required_furniture = input("Enter the furniture required: ")
required_quantity = int(input("Enter quantity of : "))
if required_furniture in furniture and required_quantity > 0:
    index = furniture.index(required_furniture)
    bill_amount = cost[index] * required_quantity
else:
    bill_amount = 0
print("Furniture Required:", required_furniture)
print("Quantity Purchased:", required_quantity)
print("Bill Amount: Rs.", bill_amount)
```

```python
#Suppressing KeyError
d={"Jhon":34, 1:23,"Harry":25, 22:'abc'}
#Accessing a key not present--suppressing
#keyError-- Method--1
print(d.get("Alice"))#returns None by default
print(d)#Alice will not be added as a key in d
##Customising get
#Override default None by Some customised value
print(d.get("Alice", "Key not found in dict"))
#Method--2
print(d.setdefault("Alice"))
#returns None and key added to
#dictionary--value None
print(d)
print(d.setdefault("Joe", 48))
#returns 48 and key added--value 48
print(d)
#{'Jhon': 34, 1: 23, 'Harry': 25, 22: 'abc',
# 'Alice': None, 'Joe': 48}
```

```python
 #sorting dictionary
d={'Jhon': 34, 'Ram': 23, 'Harry': 25, 'James': 97, 'Alice': 56, 'Joe': 48}
l=sorted(d)#returns sorted list of key
print(l)
l=sorted(d.keys())#returns list of key
print(l)
l=sorted(d.items())#returns list of sorted key value pair based on key
print(l)
l=sorted(d.values())#returns list of values in sorted form
print(l)
l=sorted(d.items(), key=lambda x:x[1])#sort dictionary on value
print(l)
l=sorted(d.items(), key=lambda x:x[0])#sort dictionary on keys
print(l)
```
